```python
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score, fbeta_score

# Load the data from the CSV file
df = pd.read_csv(r"C:\Users\haris\Desktop\imp\zzzz.masters\res-LLM\res\finalized_4_models\FINAL_BERT_RESEARCH\Bert_your_file.csv")

# Extract true labels and predicted labels
true_labels = df['True Label']
predicted_labels = df['Predicted Label']

# Calculate and print overall accuracy
overall_accuracy = (true_labels == predicted_labels).mean()
print("Overall Accuracy:", overall_accuracy)

# Unique class labels
labels = true_labels.unique()

# Initialize a list to store metrics for each class
metrics = []

# Calculate metrics for each class
for label in labels:
    # True Positives, False Positives, False Negatives for the current class
    tp = sum((true_labels == label) & (predicted_labels == label))
    fp = sum((true_labels != label) & (predicted_labels == label))
    fn = sum((true_labels == label) & (predicted_labels != label))
    tn = sum((true_labels != label) & (predicted_labels != label))

    # Calculate metrics
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    f2 = 5 * (precision * recall) / (4 * precision + recall) if (4 * precision + recall) > 0 else 0
    
    # Calculate class-specific accuracy
    class_accuracy = tp / (tp + fn) if (tp + fn) > 0 else 0

    # Store metrics for the current class
    metrics.append({
        'Class': label,
        'Accuracy': class_accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1,
        'F2 Score': f2
    })

# Convert metrics to a DataFrame for easy viewing
metrics_df = pd.DataFrame(metrics)

# Print the metrics for each class
print(metrics_df)
```
